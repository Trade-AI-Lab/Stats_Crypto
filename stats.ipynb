```python
import sys, os
from ta import trend
from src.utils.utils_df import open_dataframe, filter_dataframe_by_date_range

PATH = r"data\ETHUSDT-1m-without-aggr.zip"
df = filter_dataframe_by_date_range(
    open_dataframe(PATH=PATH).drop("unnamed: 0", axis=1),
    begin_date="2023-04-03",
    end_date="2023-05-06",
)
df.set_index('open_date', inplace=True)
print(df.head())
```

```python
hourly_df = df['open_price'].resample('H').last()
macd_hourly = trend.MACD(hourly_df).macd_diff()
df['macd_hourly'] = macd_hourly.reindex(df.index, method='ffill')
```

```python
def compute_macd(frequency:str,
                window_slow: int = 26,
                window_fast: int = 12,
                window_sign: int = 9):
    timely_df = df['open_price'].resample(frequency).last()
    macd_hourly = trend.MACD(timely_df,window_slow,window_fast,window_sign).macd_diff()
    df[f'macd_{frequency}'] = macd_hourly.reindex(df.index, method='ffill')
    print("Done")
    
compute_macd("60min")
compute_macd("15min")
compute_macd("5min")
compute_macd("3min")
compute_macd("1min")
```

```python
def compute_idles(frequency:str):
    resampled_df = df.resample(frequency).agg({
        'open_price': 'first',
        'high': 'max',
        'low': 'min',
        'close': 'last'
    })
    resampled_df.rename(columns={
        'open_price': f'open_price_{frequency}',
        'high': f'high_{frequency}',
        'low': f'low_{frequency}',
        'close': f'close_{frequency}'
    }, inplace=True)
    for col in resampled_df.columns:
        df[col] = resampled_df[col].reindex(df.index, method='ffill')
    print("Done")
```

```python
import plotly.graph_objs as go
import plotly.express as px
import plotly.subplots as sp
x = df.resample('5min').first().index
dict_Y = {
    "macd_5min":df['macd_5min'].resample('5min').first().tolist(),
    "macd_15min":df['macd_15min'].resample('5min').first().tolist(),
    "macd_1h":df['macd_60min'].resample('5min').first().tolist(),
}
fig = sp.make_subplots(rows=len(dict_Y) + 1, cols=1, shared_xaxes=True, vertical_spacing=0.05)

n = 1
for keys in list(dict_Y):
    fig.add_trace(go.Scatter(x=x.tolist(), y=dict_Y[keys], mode='lines', name=keys, showlegend=False), row=n , col=1)
    n+=1

# Update the layout
fig.update_layout(
    title="Signals",
    xaxis=dict(title="Time"),
    height=900,
)
fig.add_trace(go.Ohlc(
    x=df.resample('5min').first().index,
    open=df['open_price_5min'].resample('5min').first(),
    high=df['high_5min'].resample('5min').first(),
    low=df['low_5min'].resample('5min').first(),
    close=df['close_5min'].resample('5min').first(),
), row=len(dict_Y) + 1, col=1)

# Add titles and labels
fig.update_layout(
    title='5-Minute OHLC Data',
    xaxis_title='Time',
    yaxis_title='Price',
)
# Show the plot
fig.show("browser")
```
